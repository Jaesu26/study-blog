## 남다른 실력자 (레벨 3)

`-` 한 번에 합격을 못하고 두 번째 도전에 합격했다

`-` 첫 번째 테스트 땐 투 포인터와 이분 탐색 문제가 나왔는데 이분 탐색 문제를 제대로 못풀었다 (나는 이분 탐색이 정말 싫다..)

`-` 두 번째 테스트 땐 DP랑 시뮬레이션 문제가 나왔다. 근데 DP 문제의 마지막 효율성 테케 하나가 계속 시간 초과나서 화가 났다. 제한 시간이 얼마인지도 몰라서 계속 고통 받다가 어찌어찌 최적화하고 통과했다 

`-` 나는 레벨 2, 3이면 엄청 쉬운 거라 생각했는데 레벨 2는 무슨 4번 걸려서 합격하고 레벨 3도 한 번에 합격을 못해서 충격이 조금 크다

## 프로 코딩러 (레벨 4)

`-` 첫 번째 테스트에선 DP랑 그래프 문제가 나왔는데 그래프 문제에서 런타임 에러가 계속 발생했다. 코딩 환경이 디버깅하기에 좋지는 않았고 오류 원인도 몰라서 계속 고통만 받다가 결국 못풀었다. 근데 나중에 보니까 재귀 깊이 오류였고 재귀 깊이 제한을 늘려주니까 바로 맞혀버렸다... (나 뭐 한 거지)

`-` 진짜 호되게 당해서 앞으로 DFS 절대 안 쓰고 BFS 쓸거다. 는 불가능하고 재귀 깊이나 따박따박 늘려주자..

`-` 두 번째 테스트는 더 망했다

`-` 첫 번째 문제는 위상 정렬을 활용하면 풀 수 있을 것 같아서 바로 재귀 깊이 제한 해제하고 맞혔다. 근데 두 번째 문제를 아예 못풀었다. DP로 푸는 것 같긴 한데 정확한 접근법을 모르겠다

`-` 세 번째 테스트는 첫 번째 문제부터 답이 없었다. 문제를 보니 DP를 사용하면 될 것 같았는데 효율성 부분이 도저히 답이 없어 때려쳤다. 그 상태로 두 번째 문제는 풀지도 않은 채 빠르게 서렌쳤다

`-` 네 번째 테스트에선 아까 런타임 에러 때문에 풀지 못한 그래프 문제가 나왔다. 그래서 첫 번째 문제만 풀면 테스트를 합격할 수 있는 상황이었다. 물론 나는 첫 번째 문제를 손도 대지 못하였고 그렇게 포기를 당했다(?)

`-` 꼬라지 보니까 아는 문제로만 나오는 거 아니면 통과 못할거 같다. 그래도 여기까지 한 김에 레벨 4 테스트는 합격하고 끝내야겠다. 라고 생각해서 다섯 번째 테스트를 봤는데 빡구현 문제가 나왔고 내 집중력은 안드로메다로 떠나간 상태라 못풀고 때려쳤다

`-` 첫 번째 테스트를 통과했어야 됐는데 진짜 너무 슬프다. 1일차는 여기까지 하고 레벨 4 테스트는 다음에 통과하도록 하자
