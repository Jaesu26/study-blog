## 레벨 테스트

`-` 레벨 테스트에 나온 문제를 순서대로 정리해보자

### 입력 - 자가진단8 (해결)

- 문제 출처: [정올 516번](https://jungol.co.kr/problem/516)

`-` 기본 입출력 문제이다

### 운동회 (Sports Day) (해결)

- 문제 출처: [정올 5851번](https://jungol.co.kr/problem/5851)

`-` 인원수가 부족한 팀이 성현이의 팀이다

### 빗금채우기 (해결)

- 문제 출처: [정올 8404번](https://jungol.co.kr/problem/8404)

`-` 현재 출력해야 하는 값과 행, 열 포인터를 관리해서 풀었다. 내가 알던 브론즈 $3$치곤 매우 어렵다

### 나무꾼 미르코 (해결)

- 문제 출처: [정올 5170번](https://jungol.co.kr/problem/5170)

`-` 매개 변수 탐색으로 해결했다

### 구간의 합(PURQ) 1 (해결)

- 문제 출처: [정올 3297번](https://jungol.co.kr/problem/3297)

`-` 펜윅 트리로 쓱삭했다

### 배낭채우기4 (해결)

- 문제 출처: [정올 8944번](https://jungol.co.kr/problem/8944)

`-` 백준에 있는 [평범한 배낭 2](https://www.acmicpc.net/problem/12920) 문제와 동일하게 해결하면 된다

### 카드 배열 (포기)

- 문제 출처: [정올 2575번](https://jungol.co.kr/problem/2575)

`-` 그리디로 접근했는데 쉽지 않았다

### 전깃줄(중) (해결)

- 문제 출처: [정올 1257번](https://jungol.co.kr/problem/1257)

`-` 백준의 [전깃줄 - 2](https://www.acmicpc.net/problem/2568) 문제와 같다

### 장애물 경기 (포기) 

- 문제 출처: [정올 2994번](https://jungol.co.kr/problem/2994)

`-` 그래프를 정의한 뒤 다익스트라 알고리즘으로 최단 거리를 찾고 DFS로 역추적을 하면 될 것 같다

`-` 근데 그래프 정의를 $O\left(N^2\right)$보다 빠른 시간에 못하겠어서 포기했다
